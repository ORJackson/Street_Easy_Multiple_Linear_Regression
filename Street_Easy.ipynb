```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from mpl_toolkits.mplot3d import Axes3D

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
streeteasy = pd.read_csv('manhattan.csv')
```

```python
df = pd.DataFrame(streeteasy)
x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]
y = df[['rent']]
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)
```

```python
mlr = LinearRegression()
mlr.fit(x_train, y_train)
y_predict = mlr.predict(x_test)
```

```python
sonny_apartment = [[1, 1, 620, 16, 1, 98, 1, 0, 1, 0, 0, 1, 1, 0]] 
predict = mlr.predict(sonny_apartment)
print("Predicted rent: $%.2f" % predict)
```

Using multiple linear regression this model can predict the price of an apartment based upon its attributes.
Below the graph shows actual prices vs the predicted prices based upon the test data/

```python
plt.scatter(y_test, y_predict)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("actual rent prices vs multiple linear regression predicted prices")
plt.show()
```

printing the coefficients to see which variables carry more weight

```python
print(mlr.coef_)
```

```python
plt.scatter(df[['size_sqft']], df[['rent']], alpha=0.4)
plt.xlabel("size (sq foot)")
plt.ylabel("rent")
plt.show()
```

```python
plt.scatter(df[['min_to_subway']], df[['rent']], alpha=0.4)
plt.xlabel("minutes to subway")
plt.ylabel("rent")
plt.show()
```

```python
plt.scatter(df[['bedrooms']], df[['rent']], alpha=0.4)
plt.xlabel("bedrooms")
plt.ylabel("rent")
plt.show()
```

Using residual analysis to evaluate the accuracy of the model (1.00 is perfect, 0.7 and above is considered good)

```python
print("Train score:")
print(mlr.score(x_train, y_train))

print("Test score:")
print(mlr.score(x_test, y_test))
```

Plotting predicted rent vs actual rent

```python
plt.scatter(y_test, y_predict)
plt.plot(range(20000), range(20000))

plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Actual Rent vs Predicted Rent")

plt.show()

```
